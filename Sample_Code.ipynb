# Sample Code for the Dataset of Tensile Properties of Sub-sized Specimens

This notebook provides simple Python code for loading the dataset of tensile properties of sub-sized specimens. 

The code was developed as a supplementary material for the following manuscript that is currently under review by Scientific Data: 
* Longze Li, John W. Merickel, Yalei Tang, Rongjie Song, Joshua E. Rittenhouse, Aleksandar Vakanski, and Fei Xu, "Dataset of Tensile Properties for Sub-sized Specimens of Nuclear Structural Materials," submitted to *Scientific Data*, under review, 2024.

The dataset is stored in the Excel file titled `"Tensile_Properties_of_Sub-sized_Specimens.xlsx."` 

The code demonstrates how to load the dataset into a pandas dataframe, and also provides examples of exploratory data analysis and visualization of features in the dataset.

# 1. Loading the dataset <a name='1.-loading-the-dataset'></a>

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data as pandas dataframe
tensile_df =  pd.read_excel("Tensile_Properties_of_Sub-sized_Specimens.xlsx", header=0, skiprows=1)
tensile_df
```

```python
# Check the data information for each column
tensile_df.info()
```

# 2. Visualizing the Distribution of Features in the Dataset <a name='2.-visualizing-the-distribution-of-features-in-the-dataset'></a>

## 2.1 Sub-sized Specimen Dimensions <a name='2.1-sub-sized-specimen-dimensions'></a>

```python
# Histogram of Gauge Length of the Specimens (Figure 4a)
plt.rcParams.update({'font.size': 14.0})
fig1, ax1 = plt.subplots()
tensile_df['Gauge Length (mm)'].dropna().hist(bins=20, linewidth=0.5, color='lightgreen', edgecolor='red')
ax1.set_xlabel('Gauge Length (mm)')
ax1.set_ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
# Histogram of Thickness of the Specimens (Figure 4b)
fig2, ax2 = plt.subplots()
tensile_df['Thickness (mm)'].dropna().hist(bins=20, linewidth=0.5,color='lightgreen', edgecolor='red')
ax2.set_xlabel('Thickness (mm)')
ax2.set_ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
# Histogram of Width of the Specimens (Figure 4c)
fig3, ax3 = plt.subplots()
tensile_df['Width (mm)'].dropna().hist(bins=20, color='lightgreen',linewidth=0.5, edgecolor='red')
ax3.set_xlabel('Width (mm)')
ax3.set_ylabel('Frequency')
plt.grid(False)
plt.show()
```

## 2.2 Tensile Properties <a name='2.2-tensile-properties'></a>

```python
# Histogram of Yield Strength (Figure 5a)
bin_edges = numbers = list(range(0, 1401, 100))
fig4, ax4 = plt.subplots()
ax4.set_xlabel('Yield Strength (MPa)')
ax4.set_ylabel('Frequency')
hist, edges = np.histogram(tensile_df['Yield Strength (MPa)'].dropna(), bins=bin_edges)
bin_centers = (edges[:-1] + edges[1:]) / 2
plt.bar(bin_centers, hist, width=100, linewidth=0.5, color='skyblue', edgecolor='black') 
plt.grid(False)
plt.xlim(0, 1400)
plt.show()
```

```python
# Histogram of Ultimate Tensile Strength (Figure 5b)
bin_edges = numbers = list(range(0, 1401, 100))
fig5, ax5 = plt.subplots()
ax5.set_xlabel('Ultimate Tensile Strength (MPa)')
ax5.set_ylabel('Frequency')
hist, edges = np.histogram(tensile_df['Ultimate Tensile Strength (MPa)'].dropna(), bins=bin_edges)
bin_centers = (edges[:-1] + edges[1:]) / 2
plt.bar(bin_centers, hist, width=100, linewidth=0.5, color='skyblue', edgecolor='black') 
plt.grid(False)
plt.xlim(0, 1400)
plt.show()
```

```python
# Histogram of Uniform Elongation (Figure 5c)
bin_edges = numbers = list(range(0, 120, 5))
fig6, ax6 = plt.subplots()
ax6.set_xlabel('Uniform Elongation (%)')
ax6.set_ylabel('Frequency')
hist, edges = np.histogram(tensile_df['Uniform Elongation (%)'].dropna(), bins=bin_edges)
bin_centers = (edges[:-1] + edges[1:]) / 2
plt.bar(bin_centers, hist, width=5, linewidth=0.5, color='skyblue', edgecolor='black') 
plt.grid(False)
plt.xlim(0, 120)
plt.show()
```

```python
# Histogram of Total Elongation (Figure 5d)
bin_edges = numbers = list(range(0, 120, 5))
fig7, ax7 = plt.subplots()
ax7.set_xlabel('Total Elongation (%)')
ax7.set_ylabel('Frequency')
hist, edges = np.histogram(tensile_df['Total Elongation (%)'].dropna(), bins=bin_edges)
bin_centers = (edges[:-1] + edges[1:]) / 2
plt.bar(bin_centers, hist, width=5, linewidth=0.5, color='skyblue', edgecolor='black') 
plt.grid(False)
plt.xlim(0, 120)
plt.show()
```

## 2.3 Manufacturing Treatment and Irradiation Conditions <a name='2.3-manufacturing-treatment-and-irradiation-conditions'></a>

```python
# Histogram of Manufacturing Treatment (Figure 6a)

# Count the occurrences of each category of treatment
category_counts = tensile_df['Treatment – Three Categories'].value_counts()
# Sort the counts in ascending order
sorted_category_counts = category_counts.sort_values(ascending=True)
# Create a bar plot to visualize a sorted histogram
fig8, ax8 = plt.subplots()
ax8.bar(sorted_category_counts.index, sorted_category_counts.values, color='slategrey', linewidth=0.5, edgecolor='black', width=0.23)
ax8.set_xlabel('Manufacturing Treatment')
ax8.set_ylabel('Frequency')
ax8.grid(False)
plt.show()
```

```python
# Histogram of Irradiation Dose (Figure 6b)
fig9, ax9 = plt.subplots()
tensile_df['Irradiation Dose (dpa)'].dropna().hist(bins=10, color='slategrey',linewidth=0.5, edgecolor='black')
ax9.set_xlabel('Irradiation dose (dpa)')
ax9.set_ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
# Histogram of Irradiation Temperature (Figure 6c)
fig10, ax10 = plt.subplots()
tensile_df['Irradiation Temperature (C)'].dropna().hist(bins=10, color='slategrey',linewidth=0.5, edgecolor='black')
ax10.set_xlabel('Irradiation Temperature (C)')
ax10.set_ylabel('Frequency')
plt.grid(False)
plt.show()
```

## 2.4 Technical Validation <a name='2.4-technical-validation'></a>

```python
# Histogram of Width to Thickness Ratio (Figure 8a)
df_l_a = tensile_df[['Gauge Length (mm)', 'Thickness (mm)', 'Width (mm)']].dropna()
aspect_ratio_w_to_t = df_l_a['Width (mm)']/df_l_a['Thickness (mm)']
fig11, ax11 = plt.subplots()
aspect_ratio_w_to_t.hist(bins=20, color='darkolivegreen', edgecolor='black', linewidth = 0.5)
ax11.set_xlabel('Width-to-Thickness Ratio')
ax11.set_ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
# Histogram of Gauge Length to Square Root of the Area Ratio (Figure 8b)
aspect_ratio_l0_ao = df_l_a['Gauge Length (mm)']/np.sqrt(df_l_a['Thickness (mm)']*df_l_a['Width (mm)'])
fig12, ax12 = plt.subplots()
aspect_ratio_l0_ao.hist(bins=20, color='darkolivegreen', edgecolor='black', linewidth = 0.5)
ax12.set_xlabel('Length-to-Square_Root_of_Area Ratio')
ax12.set_ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
# Histogram of Gauge length to Width Ratio (Figure 8c)
aspect_ratio_l0_w = df_l_a['Gauge Length (mm)']/df_l_a['Width (mm)']
fig13, ax13 = plt.subplots()
aspect_ratio_l0_w.hist(bins=20, color='darkolivegreen', edgecolor='black', linewidth = 0.5)
ax13.set_xlabel('Gauge Length-to-Width Ratio')
ax13.set_ylabel('Frequency')
plt.grid(False)
plt.show()
```
